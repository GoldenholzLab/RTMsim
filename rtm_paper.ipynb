```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Low-high-testv6.csv')

df['pow_m'] *= 100
df['pow_r'] *= 100

df = df.rename(columns={'pow_m':'MPC Power %','pow_r':'RR50 Power %','minSz':'Minimum rate','baseTF':'Include baseline'})

plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
sns.scatterplot(data=df,x='N',y='MPC Power %',hue='Include baseline',style='Include baseline')
plt.title('MPC Efficacy vs. N')
plt.plot([300,1000],[90,90],'k:')
plt.ylim(0,100)
plt.grid(True)
plt.subplot(1,2,2)
sns.scatterplot(data=df,x='N',y='RR50 Power %',hue='Include baseline',style='Include baseline')
plt.plot([300,1000],[90,90],'k:')
plt.title('RR50 Efficacy vs. N')
plt.ylim(0,100)
plt.grid(True)
```

We see 
MPC, baseline false: N=350. baseline true: N=500. Here, a savings of 30%
RR50, baseline false: N=600, baseline true: N=900. Here a savings of 33%.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
d = pd.DataFrame({'MPC N':[350,500,450,550],'RR50 N':[600,900,750,1050],'Baseline Included':[False,True,False,True],
                  'Minimum rate (sz/mo)':[4,4,2,2]})
f= plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
ax = sns.barplot(data=d,x='Minimum rate (sz/mo)',y='MPC N',hue='Baseline Included')
ax.legend_.remove()
plt.grid(True)
plt.title('N for 90% power MPC')
plt.subplot(1,2,2)
sns.barplot(data=d,x='Minimum rate (sz/mo)',y='RR50 N',hue='Baseline Included')
plt.legend(bbox_to_anchor=(1.02, 0.8),title='Baseline Included')
plt.grid(True)
plt.title('N for 90% power RR50')

f.tight_layout(pad=2)
```
